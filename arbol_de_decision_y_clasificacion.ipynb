### Arbol de decision y clasificacion

*Se realizo el ejecicio con un dataset que nos provee por defecto la libreria scikit-learn*
*El dataset tiene informacion acerca de vinos y en base a ciertas condiciones que el modelo de decision predice se esta haciendo la clasificacion de los vinos*

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz
import graphviz
```

*Se realiza la carga de la informacion y se divide la data en data de prueba y data de entrenamiento*

```python
wine_dataset = load_wine()
x_ = wine_dataset.data
y_ = wine_dataset.target
x_train, x_test, y_train, y_test = train_test_split(x_, y_, test_size=0.3, random_state=14)
```

*Se crea la instancia del modelo de entrenamiento y se realiza el entrenamiento. Posterior a esto se realiza la pediccion de los datos de test*

```python
decision_model = DecisionTreeClassifier(random_state=14)
decision_model.fit(x_train, y_train)
y_pred = decision_model.predict(x_test)
```

*Una vez tenemos las predicciones que esta haciendo nuestro modelo lo que hacemos es verificar por medio de la funcion accuracy_score que tan acertado es nuestro modelo dandonos un valor entre 0 y 1. Cuando el valor tiende a 1 significa que el modelo predice y clasifica bien gran cantidad de datos*

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Precision: {accuracy}")
```

*Haciendo uso de la libreria graphviz, realizamos la representacion grafica del arbol de decision y clasificacion que automaticamente basado en datos genero nuestro modelo*

```python
dot_data = export_graphviz(
    decision_model, out_file=None,
    feature_names=wine_dataset.feature_names, class_names=wine_dataset.target_names,  
    filled=True, rounded=True, special_characters=True
)
graph = graphviz.Source(dot_data)  
graph
```
